```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
datawalk = pd.read_csv('walk.txt', sep=",", header=None)
datawalk.columns = ["x", "y", "z"]
datawalk['x'] = datawalk['x'].map(lambda x: x.lstrip('('))
datawalk['z'] = datawalk['z'].map(lambda x: x.rstrip(')'))
datawalk['Type'] = 'Walk'
datawalk.head()
```

```python
dataidle = pd.read_csv('Idle.txt', sep=",", header=None)
dataidle.columns = ["x", "y", "z"]
dataidle['x'] = dataidle['x'].map(lambda x: x.lstrip('('))
dataidle['z'] = dataidle['x'].map(lambda x: x.lstrip(')'))
dataidle['Type'] = 'Idle'
dataidle.head()
```

```python
dataJump = pd.read_csv('Jump.txt', sep=",", header=None)
dataJump.columns = ["x", "y", "z"]
dataJump['x'] = dataJump['x'].map(lambda x: x.lstrip('('))
dataJump['z'] = dataJump['z'].map(lambda x: x.rstrip(')'))
dataJump['Type'] = 'Jump'
dataJump.head()
```

```python
data = datawalk
data.append(dataidle)
data.append(dataJump)
```
